## Array Operations

```python
ray1 = np.array([1, 2, 3])
ray2 = np.array([5, 4, 3])

print("Elementwise addition", ray1+ray2)
print("Multiplication by a scalar", ray1*3)
print("Mean value", ray1.mean())
```

```python
my_list = [1, 4, 6, 9]### doesn't multiply elements rather gives the same list twice
print(my_list*2)
```

```python
# Solving system of linear equations

A = np.array([[2, 3],
              [4, 5]])

b = np.array([8, 5])

sol = np.linalg.solve(A, b)
sol
```

sol1 = np.linalg.eigvals(A)
sol1

## MATPLOTLIB

a) Line plot

```python
import matplotlib.pyplot as plt
X = [1, 2, 3, 4, 5]
Y = [3, 6, 9, 12, 15]

plt.plot(X,Y)
plt.title("Line chart")
plt.xlabel("X- Axis")
plt.ylabel("Y-Axis")
```

```python
import matplotlib.pyplot as plt
X = np.random.randn(5)
Y = np.random.randn(5)

plt.scatter(X,Y)
plt.title("Scatter Plot")
plt.xlabel("X- Axis")
plt.ylabel("Y-Axis")
```

```python
data = np.random.randn(200)

plt.hist(data, bins = 10)
plt.title("Histogram of scores")
plt.xlabel("X- Axis")
plt.ylabel("Y-Axis")
```
