````python
import math
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b**2 - (4*a*c)
print(d)
if (d == 0):   
    root0 = (-b/(2*a))
    print("root0 = ", root0)

elif d > 0:
    root1 = (-b + math.sqrt(d))/(2*a)
    root2 = (-b - math.sqrt(d))/(2*a)
    print (f"root1= {root1} and root2 = {root2} ")
else:
    print("Complex")
````

````python
sentence = "I need water for drinking."
for letter in sentence:
    if letter in "aeiou":
        print("The letter", letter, "is a vowel")
    elif letter == " ":
    
        print("This is likely a blank space")
    elif letter == (",", ".",";",":"):
        print("This is a punctuation")
        
    else:
        print(" The letter is", letter, "is a constant")
````
