## Normal equation/ Cholesky

```julia
## Write a function in Julia that solve LSQ problem by using the normal equation
A = [1 3 1; 1 3 7; 1 -1 -4; 1 -1 2]
b = [1 1 1 1]'
function least_square_problem(A,b)
    B = A'*A
    bhat = A'*b
    R = cholesky(B)
    y
    = R'\bhat
    x = R\y
    
    return x
end
```

## House holder

```julia
x = [2 -1 1]'
```

```julia
M1 = norm(x,2)
```

```julia
e = [1 0 0]'
y = M1*e
```

```julia
u = x - y
```

```julia
# Normalizing u

v = u/norm(u)
```

```julia
gamma = 2/(norm(u))^2
```

```julia
I = [1 0 0; 0 1 0; 0 0 1]
```

```julia
@time Q = I - (v*v'*gamma)
```
